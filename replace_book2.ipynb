```python
import pandas as pd
import numpy as nm

df = pd.read_excel('Book2.xlsx')
df
```

```python
new_df = df.replace(
{
    'EMP_NAME': 'no name',
    'GENDER':'nl',
    'JOB':'not specify',
    'SALARY': -99999}, nm.NaN)      # replace the dictionary of values with the NaN
new_df
```

```python
new_df = df.replace({
    
    'no name': 'kristin',
    'nl': 'null',
    'M': 'Male',
    'not specify': 'worker',
    -99999: nm.NaN
})                              #replace the values with the given values
new_df
```

```python
new_df = df.replace({
    'SALARY': '[A-Z a-z]',
    'EMP_NAME': '[0-9]'
},'',regex= True)    # replace the num, alphabets from the specific column with '' by regular_exp
new_df
```
