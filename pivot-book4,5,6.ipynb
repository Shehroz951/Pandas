```python
# pivot is used to transform or reshape the dataframe
# pivot_table is used to summarize the columns



import pandas as pd

df = pd.read_excel('Book4.xlsx')
df
```

```python
ndf = df.pivot(index='Date', columns= 'City')
ndf
```

```python
ndf = df.pivot(index='Date', columns= 'City', values = 'Weather')   # it return only weather column
ndf
```

```python
df = pd.read_excel('Book5.xlsx')
df
```

```python
ndf = df.pivot_table(index = 'City', columns = 'Date')    # it return the average of both columns
ndf
```

```python
ndf = df.pivot_table(index = 'City', columns = 'Date', aggfunc= 'sum')
ndf
```

```python
ndf = df.pivot_table(index = 'City', columns = 'Date', aggfunc = 'count')
ndf
```

```python
ndf = df.pivot_table(index = 'City', columns = 'Date', aggfunc = 'mean')
ndf
```

```python
ndf = df.pivot_table(index = 'City', columns = 'Date', margins = True)
ndf
```

```python
df = pd.read_excel('Book6.xlsx')
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])         # convert the date into datetime format
```

```python
ndf = df.pivot_table(index=pd.Grouper(freq = 'M', key = 'Date'), columns = 'City')    # freq = 'M' (months)
ndf                  # it shows the average temp and windspeed of both months
```
