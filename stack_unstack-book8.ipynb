```python
# stack is also used to transform the dataframes


import pandas as pd

df = pd.read_excel('Book8.xlsx',header= [0,1])  # header means how which rows use as a header
df
```

```python
df.stack()
```

```python
df.stack(level = 0)
```

```python
df1 = df.stack(level=1)  # by default level is 1
df1
```

```python
df2 = df1.unstack()      # unstack is used to undo the stack (back to the original form)
df2
```
