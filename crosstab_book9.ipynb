```python
import pandas as pd

df = pd.read_excel('Book9.xlsx')
df
```

```python
pd.crosstab(df.Native, df.Handed)
```

```python
pd.crosstab(df.Gender, df.Age)
```

```python
pd.crosstab(df.Gender, df.Handed)
```

```python
pd.crosstab(df.Gender, df.Handed, margins = True)
```

```python
pd.crosstab(df.Gender,[ df.Handed, df.Native], margins = True)
```

```python
pd.crosstab([df.Gender, df.Age],[ df.Handed, df.Native], margins = True)
```

```python
pd.crosstab(df.Gender,df.Handed, normalize = 'index')            # normalize shows the percentage
```

```python
import numpy as np
pd.crosstab(df.Gender, df.Handed, values= df.Age, aggfunc= np.average)
```
