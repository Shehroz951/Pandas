```python
# by xlsx (excel file)
# handle missing values by : fillna, interpolate, dropna

import pandas as pd

df = pd.read_excel('Book1.xlsx')
df.set_index('EMP_ID', inplace = True)
ndf = df.fillna(0)
ndf = df.fillna(
    {
        'EMP_ID': 0,
        'EMP_NAME': 'NO NAME',
        'GENDER': 'NOT DEFINE',
        'JOB': 'NULL',
        'SALARY': 0
    }
)
ndf = df.fillna(method = 'ffill', limit = 2) # fillna is used to fill the empty cells with the given value
ndf
ndf = df.interpolate() # interpolate is used to make a guess and fill the cell
ndf
ndf = df.dropna(thresh=3)  # dropna is used to drop the rows where any cell is empty
ndf                           # 'all' is used to drop the rows where all the cells are empty
                            # 'any' is used to drop the rows where atleast one cell is empty
                            # 'thresh = 1' is used when atleast one non-empty value
        
ndf = pd.id
```
