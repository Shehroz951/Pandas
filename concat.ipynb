```python
import pandas as pd

pk_weather = {
    'cities': ['islamabad', 'lahore', 'faisalabad'],
    'temperature': [25, 30, 28],
    'weather' : ['cloudy', 'hot day', 'partly clouded']
}
pk = pd.DataFrame(pk_weather)
pk
```

```python
aus_weather = {
    'cities' : ['sidney', 'victoria', 'melbourne'],
    'temperature': [22, 20, 25],
    'weather' : ['cloudy','stormy', 'hot day']
}
aus = pd.DataFrame(aus_weather)
aus
```

```python
df = pd.concat([pk, aus])
df
```

```python
df = pd.concat([pk, aus], ignore_index = True)
df
```

```python
df = pd.concat([pk, aus], keys = ['pakistan', 'australia'])
df
```

```python
df.loc['australia']
```

```python
windspeed_df = pd.DataFrame({
    'cities': ['multan', 'bwl', 'sialkot'],
    'windspeed': [20, 35, 30]
})
windspeed_df
```

```python
temp_df = pd.DataFrame({
    'cities': ['sialkot', 'multan'],
    'temp': [30, 40]
})
temp_df
```

```python
df = pd.concat([windspeed_df, temp_df])
df
```

```python
df = pd.concat([windspeed_df, temp_df], axis = 1)    # to concat the columns horizontaly
df
```

```python
windspeed_df = pd.DataFrame({
    'cities': ['multan', 'bwl', 'sialkot'],
    'windspeed': [20, 35, 30]
}, index = [0,1,2])         #to pass the index, store the data index wise
windspeed_df



temp_df = pd.DataFrame({
    'cities': ['sialkot', 'multan'],
    'temp': [30, 40]
}, index = [2,0])
temp_df
```

```python
df = pd.concat([windspeed_df, temp_df], axis = 1)
df
```
