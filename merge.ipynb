```python
import pandas as pd

windspeed_df = pd.DataFrame({
    'cities': ['multan', 'bwl', 'sialkot'],
    'windspeed': [20, 35, 30]
})
windspeed_df
```

```python
temp_df = pd.DataFrame({
    'cities': ['sialkot', 'multan', 'bwl'],
    'temp': [30, 40, 35]
})
temp_df
```

```python
df = pd.merge(windspeed_df, temp_df, on= 'cities')  # 'on' means on which column use to merge
df
```

```python
windspeed_df = pd.DataFrame({
    'cities': ['multan', 'bwl', 'sialkot'],
    'windspeed': [20, 35, 30]
})
windspeed_df

# change the data(insert some different values)

temp_df = pd.DataFrame({
    'cities': ['sialkot', 'multan', 'grw'],
    'temp': [30, 40, 35]
})
temp_df
```

```python
df = pd.merge(windspeed_df, temp_df, on= 'cities')   #merge is by default inner_join (return only common rows)
df
```

```python
df = pd.merge(windspeed_df, temp_df, on= 'cities', how = 'outer')  # we change the join to outer join (it is union)
df
```

```python
df = pd.merge(windspeed_df, temp_df, on= 'cities', how = 'left') # change to left join
df
```

```python
df = pd.merge(windspeed_df, temp_df, on= 'cities', how = 'right') # change to right join
df
```

```python
df = pd.merge(windspeed_df, temp_df, on= 'cities', how = 'outer', indicator = True) #indicator is used to identiy the table
df
```

```python
df1 = pd.DataFrame({
    'cities': ['multan', 'bwl', 'sialkot'],
    'windspeed': [20, 35, 30],
    'temp': [30, 40, 35]
})
        # if we use the dataframes that have the same columns
df2 = pd.DataFrame({
    'cities': ['multan', 'bwl', 'grw', 'khr'],
    'windspeed': [24, 30, 28, 36],
    'temp': [37, 42, 31, 28]
})
```

```python
df = pd.merge(df1,df2, on = 'cities')   # to identify the column it place the 'x', 'y'
df
```

```python
df = pd.merge(df1,df2, on = 'cities', how = 'outer', suffixes= ['_left', '_right']) # we can use the suffixes to identify the columns
df
```
